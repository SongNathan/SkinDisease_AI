```python
import os
```

```python
#파일 번호 순서대로 바꾸기
folder_path = r'C:\Users\yohee\SkinDisease_AI\Papular'
file_list = os.listdir(folder_path)

for idx in range(len(file_list)):
    old_name = file_list[idx]
    new_name = str(idx) + '.jpg'
    file_path = folder_path + "\\" + old_name
    new_file_path = folder_path + "\\" + new_name
    print('{} -> {}'.format(old_name, new_name))
    os.rename(file_path, new_file_path)
```

## 이미지 크기 조정
이미지 크기를 조정하는 방법에는 사용자가 원하는 값의 크기로 변환하거나, 비율에 따른 조정을 하는 방법이 있음

```python
import os
import cv2
import time
print(cv2.__version__)
```

```python
folder_path = r'C:\Users\yohee\SkinDisease_AI\Atopy' #오류발생 -> 폴더이름을 아토피 -> Atopy라고 바꿨더니 오류 사라짐
file_list = os.listdir(folder_path)

for file in file_list:
    img = cv2.imread(folder_path + "\\" + file)
    img = cv2.resize(img, (200, 190), interpolation=cv2.INTER_CUBIC)
    cv2.imshow("image", img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
